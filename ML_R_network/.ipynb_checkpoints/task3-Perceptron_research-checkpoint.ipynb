## using MNIST DATABASE Handwritten data

```python
import torch.nn as nn
import torch
import torchvision
from torch.utils.data import DataLoader
import time
```

```python


print("start")
EPOCH=10# Total number of training sessions
BATCH_SIZE=20#Batch size
LR=0.03#learning rate #cross entropy loss function
DOWNLOAD_MNIST=False# Whether to download the dataset when running the code
 
cuda_available=torch.cuda.is_available()

cuda_available=False
# Set up a transformed set that first transforms the data to 
#tensor and then normalizes it to a normal distribution with 
#a mean of .5 and a standard deviation of .5
trans=torchvision.transforms.Compose(
    [
        torchvision.transforms.ToTensor(),#ToTensor method turns [0,255] into [0,1]
        torchvision.transforms.Normalize([0.5],[0.5])#mean (mean), std (standard deviation standard deviation)
    ]
)
 
print("load data")
train_data=torchvision.datasets.MNIST(
    root="./mnist",# Set the root directory of the dataset
    train=True,#Whether it is a training set
    transform=trans,#Conversion of data
    download=DOWNLOAD_MNIST
)

train_loader=DataLoader(train_data,batch_size=BATCH_SIZE,shuffle=True)
 
test_data=torchvision.datasets.MNIST(
    root="./mnist",
    train=False,
    transform=trans,
    download=DOWNLOAD_MNIST
)
test_loader=DataLoader(test_data,batch_size=len(test_data),shuffle=False)
print("net creating")
 

net=torch.nn.Sequential(
    nn.Linear(28*28,30),#28*28 inputs, 30 outputs
    nn.Tanh(),
    nn.Linear(30,10)#30 inputs, 10 outputs, corresponding to 10 categories
)
 

```

```python
for data in train_loader:
    img,label=data
    print(img.shape)
```

```python
if cuda_available:
    net.cuda()
#Define loss functions and optimization methods
loss_function=nn.CrossEntropyLoss()#cross-entropy loss function
optimizer=torch.optim.SGD(net.parameters(),lr=LR)#Optimization methods
 
print("start training")
for ep in range(EPOCH):
    # Record how long it takes to train + test all the datasets once
    startTick = time.clock()
    for data in train_loader:#For each batch of the training set
        img,label=data
        img = img.view(img.size(0), -1)# Flattening the image into a one-dimensional vector
 
        if cuda_available:
            img=img.cuda()
            label=label.cuda()
 
        out=net(img)#Send into the network for output
        loss=loss_function(out,label)#GettingLoss
 
        optimizer.zero_grad()#gradient zeroing
        loss.backward()# Backpropagation to get the gradient, but the parameters have not been updated
        optimizer.step()#Update gradient
 
    num_correct=0# of correct classifications, test accuracy in test set
    #Because the batchsize of the test set is the length of the test set, the following loop is only once
    for data in test_loader:
        img,label=data
        img = img.view(img.size(0), -1)
 
        if cuda_available:
            img=img.cuda()
            label=label.cuda()
 
        out=net(img)#Get Output
 
        _,prediction=torch.max(out,1)
        print(prediction)
        #torch.max() returns two results.
        # the first being the maximum value and the second being the corresponding index value.
        # The second argument 0 means take the maximum value by column and return the corresponding row index value, 
        #1 means take the maximum value by row and return the corresponding column index value.
        num_correct+=(prediction==label).sum()#Find the quantity for which the predicted and true values are the same, i.e., the quantity for which the prediction is correct
 
    accuracy=num_correct.cpu().numpy()/len(test_data)
    timeSpan=time.clock()-startTick
    print("The %dth iteration period, the accuracy is %f,time %dS"%(ep+1,accuracy,timeSpan))
    torch.save(net.state_dict(), './params.pth')
    
```
